### Housing Case Study

**Problem Statement**

Optimising sale prices of properties in a city by a real estate company based on factors such as area, bedrooms, parking etc

Goals:

1. Identify variables affecting house prices.
2. Create a linear model that quantitavely relates house prices with the variables
3. Determine the accuracy of the model

#### Importing and understanding Data

```python
import pandas as pd

housing = pd.read_csv('Housing.csv')

housing.head()
```

```python
housing.info()
```

```python
housing.describe()
```

#### Data Preparation

```python
# COnvert yes to 1 and No to 0
housing['mainroad'] = housing['mainroad'].map({'yes': 1, 'no' : 0 })
housing['basement'] = housing['basement'].map({'yes': 1, 'no' : 0 })
housing['guestroom'] = housing['guestroom'].map({'yes': 1, 'no' : 0 })
housing['hotwaterheating'] = housing['hotwaterheating'].map({'yes': 1, 'no' : 0 })
housing['airconditioning'] = housing['airconditioning'].map({'yes': 1, 'no' : 0 })
housing['prefarea'] = housing['prefarea'].map({'yes': 1, 'no' : 0 })

housing.head()
```

```python
# furnishing status has 3 options. we need to convert them to integer values
# creating dummy variables for furnishingstatus

status = pd.get_dummies(housing['furnishingstatus'])

status.head()
```

```python
# we only need 2 columns in status
# we drop first column
status = pd.get_dummies(housing['furnishingstatus'], drop_first=True)

status.head()
```

```python
# adding the result to the neater dataframe
housing_new = pd.concat([housing, status], axis=1)
housing_new.head()
```

```python
# drop the furnisgingstatus column
housing_new.drop(['furnishingstatus'], axis=1, inplace=True)
housing_new.head()
```

```python
# adding new variables that may help us. If they dont we can remove them later

# adding bathrooms per bedroom
housing_new['bbratio'] = housing_new['bathrooms'] / housing_new['bedrooms']

# adding area and bedroom ratio
housing_new['areaperbedroom'] = housing_new['area'] / housing_new['bedrooms']

housing_new.head()
```

#### Re-Scaling features

```python
# we use normalization technique
import numpy as np

def normalize(x):
    return ((x - np.min(x)) / (max(x) - min(x)) )

# apply the normalize function
housing_new = housing_new.apply(normalize)
housing_new.head()
```

#### Split data to training and testing datasets

```python
housing_new.columns
```

```python
# create feature variable X
X = housing_new[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'semi-furnished', 'unfurnished', 'bbratio',
       'areaperbedroom']]

# create response varaible (label)
y = housing_new['price']

# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

#### Building a linear model with statsmodels

```python
import statsmodels.api as sm

# adding a constant column to dataframe
X_train = sm.add_constant(X_train)

# create first fitted model
lm = sm.OLS(y_train, X_train).fit()

# check summarry
lm.summary()

```

#### Handling multi-correlinearity problem (Variance Inflation Factor)

```python
 # VIF Calculator function
def vif_calc(input_data, dependent_col):
    vif_df = pd.DataFrame(columns = ['Var', 'Vif'])
    x_vars = input_data.drop([dependent_col], axis=1)
    xvar_names = x_vars.columns
    for i in range(0, xvar_names.shape[0]):
        y = x_vars[xvar_names[i]]
        x = x_vars[xvar_names.drop(xvar_names[i])]
        r_squared = sm.OLS(y,x).fit().rsquared
        vif = round(1 / (1 - r_squared), 2)
        vif_df.loc[i] = [xvar_names[i], vif]
    return vif_df.sort_values(by= 'Vif', axis=0, ascending=False, inplace=False)


# calculating VIF value
vif_calc(input_data=housing_new, dependent_col='price')
```

**we can see that bbratio has a high VIF and a high p value. we will therefore drop the column**

```python
# before deleting, lets first see the correlation matrix heatmap
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(16,12))
sns.heatmap(housing_new.corr(), annot=True)
```

#### Dropping the Variable and Updating the model

```python
X_train = X_train.drop('bbratio', 1)

# # create a second fitted model
lm_2 = sm.OLS(y_train, X_train).fit()

# # show summary
lm_2.summary()
```

```python
# recalcculating VIF and inspecting the values again
vif_calc(input_data=housing_new.drop(['bbratio'], 1), dependent_col='price')
```

**bedroom values seem to be higher, we will drop it**

```python
X_train = X_train.drop('bedrooms', 1)

lm_2 = sm.OLS(y_train, X_train).fit()

lm_2.summary()
```

```python
# recalcculating VIF and inspecting the values again
vif_calc(input_data=housing_new.drop(['bbratio', 'bedrooms'], 1), dependent_col='price')
```

```python
# removing areaperbedroom
X_train = X_train.drop('areaperbedroom', 1)

lm_2 = sm.OLS(y_train, X_train).fit()

lm_2.summary()
```

```python
# recalcculating VIF and inspecting the values again
vif_calc(input_data=housing_new.drop(['bbratio', 'bedrooms', 'areaperbedroom'], 1), dependent_col='price')
```

```python
# semifutnished p value is still very high. Its not improving. we will remove it
# removing areaperbedroom
X_train = X_train.drop('semi-furnished', 1)

lm_2 = sm.OLS(y_train, X_train).fit()

lm_2.summary()
```

```python
# recalcculating VIF and inspecting the values again
vif_calc(input_data=housing_new.drop(['bbratio', 'bedrooms', 'areaperbedroom', 'semi-furnished'], 1), dependent_col='price')
```

**the values are improving. In case you are wondering, Im trying to bring all p values to 0.02 or less and VIF to less than 5**

```python
# removing basement
X_train = X_train.drop('basement', 1)

lm_2 = sm.OLS(y_train, X_train).fit()

lm_2.summary()
```

```python
# recalcculating VIF and inspecting the values again
vif_calc(input_data=housing_new.drop(['bbratio', 'bedrooms', 'areaperbedroom', 'semi-furnished', 'basement'], 1), dependent_col='price')
```

**The values are now good enough**

## Making Predictions Using Final Model

```python
X_test_final = sm.add_constant(X_test)

# dropping some columns
X_test_final = X_test_final.drop(['bbratio', 'bedrooms', 'areaperbedroom', 'semi-furnished', 'basement'], axis=1)

# perfoming the prediction
y_pred_final = lm_2.predict(X_test_final)
```

## Model Evaluation

```python
# drawing a line graph of actual vs predicted values
c = [i for i in range(1, 165, 1) ]
fig = plt.figure(figsize=(10,8))
plt.plot(c, y_test, color='blue', linewidth=2.5) # actual values
plt.plot(c, y_pred_final, color='red', linewidth=2.5) # predicted values
fig.suptitle('Actual and Predicted Values')
plt.xlabel('index')
plt.ylabel('Housing Price')
```

```python
# using scatter plot to check the correlation of y_test and y_pred
fig = plt.figure(figsize=(10,8))
plt.scatter(y_test,y_pred_final)
fig.suptitle('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```

```python
# calculating the root mean squared
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, y_pred_final))
```
