```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)

plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

```python
# Subplots
x = np.linspace(1, 10, 100)
y = np.log(x)

# initiate a new figure
plt.figure(1)

# create subplot with 1 row 2 columns
# create first subplot in figure 1
plt.subplot(121) # equivalent tp plt.subplot(1,2,1)
plt.title(" y = log(x)")
plt.plot(x,y)

# create the second subplot in figure 1
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)

plt.show()
```

```python
# 4 subplots
x = np.linspace(1, 10, 100)

# initiate a new figure
plt.figure(1)

# subplot 1
plt.subplot(2, 2, 1)
plt.title('Linear')
plt.plot(x, x)

# subplot 2
plt.subplot(2, 2, 2)
plt.title('Cubic')
plt.plot(x, x**3) 

# subplot 3
plt.subplot(2, 2, 3)
plt.title('Cubic')
plt.plot(x, np.log(x))

# subplot 4
plt.subplot(2, 2, 4)
plt.title('Exponential')
plt.plot(x, x**2)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.show()
```

```python
# common plots
import pandas as pd
df = pd.read_csv('market_fact.csv')
df.head()
```

```python
 # boxplot: visualize the distribution of a continous variable
plt.boxplot(df['Order_Quantity'])
plt.show()
```

```python
#sales boxplot
# its quite unreadable since Sales varies accross a wide range
plt.boxplot(df['Sales'])
plt.show()
```

```python
df['Sales'].describe()
```

```python
# solution is to change scale of y axis
# transform it to a logarithmic scale
# plotting them side by side
plt.subplot(1, 2, 1)
plt.title("Normal scale")
plt.boxplot(df['Sales'])

# logarithmic scale
plt.subplot(1, 2, 2)
plt.title("Log scale")
plt.boxplot(df['Sales'])
plt.yscale('log')

plt.show()
```

```python
# check distribution of a single variable using a histogram
plt.hist(df['Sales'])
plt.show()
```

```python
# we can transform y scalr to get better view
plt.hist(df['Sales'], log=True)
plt.show()
```

```python
# scatter plot to visualize 2 variables, one on each axis
plt.scatter(df['Sales'], df['Profit'])
plt.show()
```
